```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ast

# Load data
df = pd.read_csv("movies.csv")

# Helper to count list-like items
def safe_count(val):
    if pd.isnull(val):
        return 0
    try:
        return len(ast.literal_eval(val))
    except:
        return 0

# Feature Engineering
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year
df['title_length'] = df['title'].fillna('').apply(len)
df['num_genres'] = df['genres'].apply(safe_count)
df['num_keywords'] = df['keywords'].apply(safe_count)
df['num_production_companies'] = df['production_companies'].apply(safe_count)
df['overview_length'] = df['overview'].fillna('').apply(lambda x: len(x.split()))
df['has_tagline'] = df['tagline'].notnull().astype(int)
df['profit'] = df['revenue'] - df['budget']
df['roi'] = np.where(df['budget'] != 0, df['revenue'] / df['budget'], np.nan)
df['log_budget'] = np.log1p(df['budget'])
df['log_revenue'] = np.log1p(df['revenue'])

# Select features for correlation
features = [
    'budget', 'log_budget', 'revenue', 'log_revenue', 'profit', 'roi',
    'popularity', 'runtime', 'vote_average', 'vote_count',
    'has_tagline', 'release_year',
    'num_keywords', 'num_genres', 'num_production_companies', 'overview_length', 'title_length'
]

# Drop rows with missing values in selected features
df_corr = df[features].dropna()

# Compute correlation matrix
corr_matrix = df_corr.corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, cbar=True)
plt.title("Enhanced Correlation Heatmap for Movie Dataset", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#THIS FUNCTION IS USED TO CONVERT THE LIST OF DICTIONARY TO DICTIONARY
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# THIS FUNCTION REMOVES ALL SPACES BETWEEN STRINGS
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
import pandas as pd
import ast
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from scipy.sparse import hstack

# Load datasets
movies_df = pd.read_csv('movies.csv')
credits_df = pd.read_csv('credits.csv')

# Merge using correct keys
movies_df = movies_df.merge(credits_df, left_on='id', right_on='movie_id')

# Extract 'name' fields from JSON strings
def extract_names(text, key='name'):
    try:
        items = ast.literal_eval(text)
        return ' '.join([item[key] for item in items if key in item])
    except:
        return ''

# Apply extraction
movies_df['genres'] = movies_df['genres'].apply(lambda x: extract_names(x))
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: extract_names(x))
movies_df['cast'] = movies_df['cast'].apply(lambda x: extract_names(x))
movies_df['crew'] = movies_df['crew'].apply(lambda x: extract_names(x))

# Combine features for clustering
movies_df['combined_features'] = (
    movies_df['genres'] + ' ' +
    movies_df['keywords'] + ' ' +
    movies_df['cast'] + ' ' +
    movies_df['crew']
)

# Hybrid vectorization (TF-IDF + CountVectorizer)
vectorizer_tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
vectorizer_count = CountVectorizer(stop_words='english', max_features=5000)

X_tfidf = vectorizer_tfidf.fit_transform(movies_df['combined_features'])
X_count = vectorizer_count.fit_transform(movies_df['combined_features'])

# Merge features
X = hstack([X_tfidf, X_count])

# Dimensionality Reduction (SVD with 300 components)
svd = TruncatedSVD(n_components=300, random_state=42)
X_reduced = svd.fit_transform(X)

# Silhouette score for different k
scores = []
clusters_range = list(range(2, 21))
for k in clusters_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10, algorithm='elkan')
    labels = km.fit_predict(X_reduced)
    score = silhouette_score(X_reduced, labels)
    scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(clusters_range, scores, marker='o', linestyle='-', color='green')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('KMeans Clustering: Optimal k using Silhouette Score')
plt.grid(True)
plt.show()

# Find best k and refit model
best_k = clusters_range[scores.index(max(scores))]
print(f'Best k: {best_k} with Silhouette Score: {max(scores):.4f}')

# Final clustering with best k using Elkan algorithm (optimized for sparse data)
final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10, algorithm='elkan')
movies_df['cluster'] = final_kmeans.fit_predict(X_reduced)
```
